```python
#module import


from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn import model_selection
from sklearn import svm


digits = load_digits()
digits_data = digits.data
digits_label = digits.target

print("target_names:",digits.target_names,'\n\n\n\n',''' "DESCR:",digits.DESCR,'\n\n\n\n' ''')

X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.4, 
                                                    random_state=32)


decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print("decision_tree : \n\n",classification_report(y_test, y_pred),'\n\n')
print("A_accuacy :", accuracy_score(y_test,y_pred),'\n\n')




random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print("RandomForest: \n\n",classification_report(y_test, y_pred),'\n\n')
print("B_accuacy :", accuracy_score(y_test,y_pred),'\n\n')


svm_model = svm.SVC()


svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print("svm: \n\n",classification_report(y_test, y_pred),'\n\n')
print("C_accuacy :", accuracy_score(y_test,y_pred),'\n\n')


sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print("sgd: \n\n",classification_report(y_test, y_pred),'\n\n')
print("D_accuacy :", accuracy_score(y_test,y_pred),'\n\n')
'''
# from sklearn.linear_model import LogisticRegression
# from sklearn.pipeline import make_pipeline
# from sklearn.preprocessing import StandardScaler

# logistic_model = LogisticRegression()
# logistic_model.fit(X_train, y_train)
# y_pred = logistic_model.predict(X_test)

iteration max error occure. so, use pipeline 
'''


pipe = make_pipeline(StandardScaler(), LogisticRegression())
pipe.fit(X_train, y_train)  # apply scaling on training data

pipe.score(X_test, y_test)  # apply scaling on testing data, without leaking training data.


print("Logistic_Regression: \n\n",classification_report(y_test, y_pred))
print("E_accuacy :", accuracy_score(y_test,y_pred),'\n\n')




```

# pipeline이란?


위 사진처럼 digits dataset 코드를 실행시켜보면 결과값은 출력이 되나 f-score 관련 warning 문구가 발생한다.
pipelin은 sklearn에서 제공하는 기능으로서 데이터 전처리 과정에서 발생하는 불확실성을 줄여주고, 
데이터가 들어가는 통로를 확실하게 만들어준다는 개념으로서 실수를 사전에 예방할 수 있다.
따라서 sklearn 에서 제공하는 pipeline을 사용해 warning 문구를 제거할 수 있다.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.svm import SVC
import numpy as np

dataset = datasets.load_digits()
X = dataset.data
y = dataset.target
print('X.shape:', X.shape)
print('y.shape:', y.shape)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.4,
    random_state=32
)

models = []
models.append(('dc', DecisionTreeClassifier()))
models.append(('lr', LogisticRegression(max_iter=10000)))
models.append(('rf', RandomForestClassifier()))
models.append(('svm', SVC()))
models.append(('sgd', SGDClassifier()))

results = []
names = []
scoring = 'accuracy'
for name, model in models:
    score = model_selection.cross_val_score(model, X_train, y_train, scoring='accuracy', cv=5)
    results.append(score.mean())
    names.append(name)
    a = "%s: %f (%f)" % (name, score.mean(), score.std())
    # print(name, score)

print(results)
# print(y.target_names)


x = np.arange(5)
Y = np.array(results)


plt.xticks(ticks=[0,1,2,3,4], labels=['dc','lr','rf','svm','sgd'])
plt.bar(x,Y,width=0.4)
plt.title('Compare models')
plt.ylabel('Accuracy')
plt.show()
```

# cross validation
데이터를 학습,테스트 용도로 분리할수 없을 정도로 데이터 양이 적을때 cross validation기법을 사용한다.
cv= 5로 지정하면 학습데이터를 5개로 나누어서 한개씩 뺀 뒤 학습시키고,빼놓은 그룹으로 성능을 평가한다.
따라서 총 5개의 평가 결과가 나오는데 이 평가 결과의 평균을 성능으로 사용한다.
